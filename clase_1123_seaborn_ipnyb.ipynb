# **Distribuciones**

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins
```

```python
sns.displot(penguins, x = "flipper_length_mm")
```

```python
sns.displot(penguins, x = "flipper_length_mm", binwidth = 3)
#separacin entre valores d 3
```

```python
sns.displot(penguins, x = "flipper_length_mm", bins = 20)
#numero de columnas
```

```python
sns.displot(penguins, x = "flipper_length_mm", hue = 'species')
#dividimos por categoras
```

```python
sns.displot(penguins, x = "flipper_length_mm", hue = 'species', element = "step")
#areas continuas
```

```python
sns.displot(penguins, x = "flipper_length_mm", hue = 'species', multiple = "stack")
#evitamos reas superpuestas
```

```python
sns.displot(penguins, x = "flipper_length_mm", hue = 'species', stat = "density")
#normalizacin de base por altura de cada especie igual a 1
```

```python
sns.displot(penguins, x = "flipper_length_mm", hue = 'species', stat = "probability")
#normalizacin en tanto por 1, un poco ms e 16% de los pinguinos Adeile tienen altura de 190
```

```python
sns.displot(penguins, x = "flipper_length_mm", hue = 'species', stat = "probability", common_norm = False)
#Probabilidad sobre cada una de las especies
```

# Kernel Density **Estimation**

```python
sns.displot(penguins, x = "flipper_length_mm", kind = "kde")
```

```python
sns.displot(penguins, x = "flipper_length_mm", kind = "kde", bw_adjust= .25)
#suavizamos menos las curvas
```

```python
sns.displot(penguins, x = "flipper_length_mm", kind = "kde", bw_adjust= 2)
#suavizamos ms las curvas
```

```python
sns.displot(penguins, x = "flipper_length_mm", kind = "kde", bw_adjust= 10)
#suavizamos ms las curvas
```

```python
sns.displot(penguins, x = "flipper_length_mm", kind = "kde",hue= "species" )
```

```python
sns.displot(penguins, x = "flipper_length_mm", kind = "kde",hue= "species", fill = True )
```

# **Barplots**

```python
titanic = sns.load_dataset("titanic")
```

```python
sns.catplot(x="deck", kind="count", palette= "ch:.25", data=titanic)
```
